```python
import pandas as pd
import time
from astroquery.mast import Catalogs
from collections import defaultdict
```

```python
blcat = pd.read_excel('data/combined_dataset_BLCat.xlsx')
```

```python
def ra_to_deg(x):
    hms = x.split()
    return (int(hms[0])+(int(hms[1])+float(hms[2])/60)/60) * 15
    
    
def del_to_deg(x):
    dms = x.split()
    if dms[0][0] == '-':
        return int(dms[0]) - (int(dms[1])+float(dms[2])/60)/60
    else:
        return int(dms[0]) + (int(dms[1])+float(dms[2])/60)/60


blcat['RA_deg'] = blcat['RA (J2000.0)'].apply(ra_to_deg)
blcat['Dec_deg'] = blcat['Dec (J2000.0)'].apply(del_to_deg)
```

```python
result = defaultdict(list)

j, n = 0, blcat.shape[0]
blcat.reset_index(inplace=True, drop=True)
for index, obj in blcat.iterrows():
    j += 1
    print(f'{j} out of {n}')
    coords = str(obj['RA_deg']) + ' ' + str(obj['Dec_deg'])
    response = Catalogs.query_object(coords, radius=.002, catalog='Galex')
    if len(response) > 0:
        result['BZCAT5 Source name'].append(obj['BZCAT5 Source name'])
        result['GALEX_offset'].append(response['distance_arcmin'][0] * 60) 
        result['fuv_mag'].append(response['fuv_mag'][0])
        result['fuv_magerr'].append(response['fuv_magerr'][0])
        result['nuv_mag'].append(response['nuv_mag'][0])
        result['nuv_magerr'].append(response['nuv_magerr'][0])
        result['fuv_flux'].append(response['fuv_flux'][0])
        result['fuv_fluxerr'].append(response['fuv_fluxerr'][0])
        result['nuv_flux'].append(response['nuv_flux'][0])
        result['nuv_fluxerr'].append(response['nuv_fluxerr'][0])
        result['e_bv'].append(response['e_bv'][0])
        time.sleep(0.2)

result_df = pd.DataFrame(result)
result_df.to_excel('data/GALEX_data.xlsx')
```
