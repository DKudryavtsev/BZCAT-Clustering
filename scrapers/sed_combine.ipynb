```python
import pandas as pd
import os
from collections import defaultdict
```

```python
sed_dir = '/home/dkudr/Yandex.Disk/Development/BZCAT-Clustering/data/seds'
os.chdir(sed_dir)
files = os.listdir(sed_dir)
sed = defaultdict(list)
non_residents = [
    'BSDC-MAGIC', 'BSDC-Veritas', 'Catalina', 'NED', 
    'VizieR', '2MASS', 'SDSS13', 'SDSS7', 'UKIDSS8', 'USNO']

for file in files:
    with open(file, 'r') as f:
        ra = f.readline().split()
        source = ra[3] + ra[4] + ra[5][:2]
        dec = f.readline().split()
        source += dec[3] + dec[4] + dec[5][:2]
        ignore, count = False, 0
    
        for line in f:
            vals = line.split()
            
            # Exclude non-residents
            if (line[0] == '#') and (count > 0):
                ignore = False
            if vals[1] in non_residents:
                ignore = True
                count = 0
                           
            if (line[0] != '#'):
                count += 1  
                if not ignore:
                    freq, bin, nufnu, nufnu_err, tstart, tstop \
                        = float(vals[0]), float(vals[1]), float(vals[2]), \
                          float(vals[3]), float(vals[4]), float(vals[5])
                    comment = ''
                    if len(vals) > 6:
                        for i in range(7, len(vals)):
                            comment += vals[i] + ' '
                        comment = comment[:-1]
                    sed['Source'].append(source)
                    sed['Frequency (Hz)'].append(freq)
                    sed['Bin'].append(bin)
                    sed['Nufnu (erg cm^-2 s^-1)'].append(nufnu)
                    sed['Nufnu_err'].append(nufnu_err)
                    sed['TStart'].append(tstart)
                    sed['TStop'].append(tstop)
                    sed['Comment'].append(comment)
                        
sed_df = pd.DataFrame(sed) \
    .sort_values(['Source', 'Frequency (Hz)']).reset_index(drop=True)
sed_df.to_csv('/home/dkudr/Yandex.Disk/Development/BZCAT-Clustering/data/seds_residents.csv', 
              index=False)        
```
